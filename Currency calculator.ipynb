# Conventional Currency Calculator

```python
from tkinter import *
window=Tk()
window.title("Currency calculator")

label=Label(text="Foreign Currency",font=("Futura",8),fg="black")
label.grid(row=0,column=0)
label2=Label(text="Indian Currency",font=("Futura",8),fg="black")
label2.grid(row=1,column=0)

e=Entry(width=30,borderwidth=7,bg="light grey")
e.grid(row=0,column=1,columnspan=2)

f=Entry(width=30,borderwidth=7,bg="light grey")
f.grid(row=1,column=1,columnspan=2)

def buttonClick(number):
    global value
    value=e.get()
    e.delete(0,END)
    e.insert(0,str(value)+str(number))

def buttonClear():
    e.delete(0,END)
    f.delete(0,END)

def buttonKwd():
    global num
    num=int(e.get())
    f.insert(1,round(num*272.74,2))

def buttonEur():
    num1=int(e.get())
    f.insert(1,round(num1*89.80,2))

def buttonUsd():
    num2=int(e.get())
    f.insert(1,round(num2*83.62,2))


def buttonGbp():
    num3=int(e.get())
    f.insert(1,round(num3*106.47,2))

def buttonBhd():
    num4=int(e.get())
    f.insert(1,round(num4*221.74,2))

def buttonOmr():
    num5=int(e.get())
    f.insert(1,round(num5*217.12,2))

def buttonChf():
    num6=int(e.get())
    f.insert(1,round(num6*93.11,2))

def buttonKyd():
    num7=int(e.get())
    f.insert(1,round(num7*100.23,2))




button1=Button(text=1,padx=40,pady=20,bg="grey",fg="white",command=lambda:buttonClick(1)).grid(row=3,column=0)
button2=Button(text=2,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(2)).grid(row=3,column=1)
button3=Button(text=3,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(3)).grid(row=3,column=2)
button4=Button(text=4,padx=40,pady=20,bg="grey",fg="white",command=lambda:buttonClick(4)).grid(row=4,column=0)
button4=Button(text=4,padx=40,pady=20,bg="grey",fg="white",command=lambda:buttonClick(4)).grid(row=4,column=0)
button5=Button(text=5,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(5)).grid(row=4,column=1)
button6=Button(text=6,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(6)).grid(row=4,column=2)
button7=Button(text=7,padx=40,pady=20,bg="grey",fg="white",command=lambda:buttonClick(7)).grid(row=5,column=0)
button8=Button(text=8,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(8)).grid(row=5,column=1)
button9=Button(text=9,padx=41,pady=20,bg="grey",fg="white",command=lambda:buttonClick(9)).grid(row=5,column=2)
button0=Button(text=0,padx=40,pady=20,bg="grey",fg="white",command=lambda:buttonClick(0)).grid(row=6,column=0)
buttonkwd=Button(text="KWD",padx=30,pady=20,bg="light grey",fg="black",command=buttonKwd).grid(row=2,column=0)
buttoneur=Button(text="EUR",padx=33,pady=20,bg="light grey",fg="black",command=buttonEur).grid(row=2,column=1)
buttonusd=Button(text="USD",padx=33,pady=20,bg="light grey",fg="black",command=buttonUsd).grid(row=2,column=2)
buttongbp=Button(text="GBP",padx=30,pady=20,bg="orange",fg="white",command=buttonGbp).grid(row=2,column=3)
buttonbhd=Button(text="BHD",padx=30,pady=20,bg="orange",fg="white",command=buttonBhd).grid(row=3,column=3)
buttonomr=Button(text="OMR",padx=28,pady=20,bg="orange",fg="white",command=buttonOmr).grid(row=4,column=3)
buttonchf=Button(text="CHF",padx=30,pady=20,bg="orange",fg="white",command=buttonChf).grid(row=5,column=3)
buttonkyd=Button(text="KYD",padx=30,pady=20,bg="orange",fg="white",command=buttonKyd).grid(row=6,column=3)
buttonclear=Button(text="Clear",padx=80,pady=20,bg="light grey",fg="black",command=buttonClear).grid(row=6,column=1,columnspan=2)

window.mainloop()
```
